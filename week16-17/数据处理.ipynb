```python
#{月份:{mm-dd:用电量}}的字典,根据元素的下标+1来确定日期
data = {}
for i in df.columns[1:13]:
    count = 1
    data[i] = {}
    for j in range(len(df[i])):
        #每遍历到新的月份就+1
        if j%31 == 0:
            count += 1
        data[i][str(count)+'-'+str(j%31+1)] = df[i][j]


data

```

fimport pandas as pd

# 读取Excel文件
df = pd.read_excel('副本2022送电量.xlsx')

# 使用melt()函数将宽表格转换为长表格
df_melted = pd.melt(df, var_name="Month", value_name="Electricity Usage")

# 拆分"Month"列为"Month"和"Day"两列
df_melted[["Month", "Day"]] = df_melted["Month"].str.split("月", expand=True)
df_melted["Month"] = pd.to_numeric(df_melted["Month"])

# 创建日期列
df_melted["Date"] = df_melted.apply(lambda row: "{:02d}-{:02d}".format(row["Month"], row["Day"]), axis=1)

# 将"Date"列设置为索引
df_melted.set_index("Date", inplace=True)

# 删除没用的列
df_melted = df_melted[["Electricity Usage"]]

# 输出结果
print(df_melted)
